```python
from crewai import LLM

llm = LLM(

    model = "openrouter/mistralai/mistral-7b-instruct",
    temperature = 0.1
)

try:
    response = llm.call("Who invented transcendental meditation.")
    print("LLM Test Response:", response)
    
except Exception as e:
    print(f"Error: {e}")
    print("\nTo fix this error:")
    print("1. Get your API key from: https://openrouter.ai/keys")
    print("2. Replace 'your_openrouter_api_key_here' with your actual API key")
    print("3. Or create a .env file with: OPENROUTER_API_KEY=your_actual_key")
```
